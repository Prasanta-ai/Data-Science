# Permutations and Count

```python
from itertools import permutations
count = 0
data = [1, 2, 3]
perm = permutations(data, 2)

for i in perm:
  print(i)
  count += 1

print("Permutations Count:", count)
```

# Combinations and Count

```python
from itertools import combinations
count = 0
data = [1, 2, 3]
perm = combinations(data, 2)

for i in perm:
  print(i)
  count += 1

print("Combinations Count:", count)
```

```python
from itertools import combinations_with_replacement as cr
count = 0
data = [1, 2, 3]
perm = cr(data, 2)

for i in perm:
  print(i)
  count += 1

print("Permutations Count:", count)
```
