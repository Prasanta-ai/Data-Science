## Conditional Statement
- the if else statement

```python
x = 5
y = 2

if y < 6:
  z = x / y
  print("Make sense " ,z)
```

```python
if 5 > 6:
  print("No")
```

- The above statement printed nothing. Which means 5 is not greater than 6.
- What if a condition is false and we need to print it out?

```python
if 10 % 100 == 0:
  print("Yes")
else:
  print("No")
```

- Multipal if statements

```python
if 9 > 2:
  print("1")
if 6 > 3:
  print("2")
if 8 > 6:
  print("3")
```

- elif statement

```python
x = 7
y = -1
if x > y:
  print("Greater")
elif x < y:
  print("Smaller")
elif x == y:
  print("Equal")
elif x < 0:
  print("Negative")
```

**Note:** elif satatements are used where only one condition can be true. The rest are discarted.
