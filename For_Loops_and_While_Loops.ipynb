# While

```python
i = 0

while i != 10:
  print("Loop 4")
  i = i + 1
```

# For Loops

```python
for x in range(3):
  print("d")
```

## Looping lists

```python
my_list = [1, 2, 3, 4, 'python', 'is', 'neat']


for item in my_list:
  print(item)
```

## break
- Stop the execution of the loop.

```python
for item in my_list:
  if item == 'python':
    break
  print(item)
```

## continu
- Continue to the next item without executing the lines occuring after continue inside the loop.

```python
for item in my_list:
  if item == 2:
    continue
  print(item)
```

## enumerate()
- In case you need to also know the index.

```python
for idx, val in enumerate(my_list):
  print('idx: {}, value: {}'.format(idx, val))
```
