## String Indexing

- String can be stored into variables.
- Each character can be accessrd seperately from a string.


```python
greetings = "Hello World!"

print(greetings)

first = "Hello "
second = "World!"

join = first + second

print(join)
```

```python
join[4]
```

```python
print("First Character: ", join[0])
print("Second Character: ", join[1])
print("Last character: ", join[11])
```

## String Slicing
- Extracting a perticular substring from the string.
- some_string[start:end]

```python
greetings = "Hello World!"

print("First part --> ", greetings[0:5])

print("Second part --> ", greetings[6:12])
```

```python
print(greetings[0:1])
```
